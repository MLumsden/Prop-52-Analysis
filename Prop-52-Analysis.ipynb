# Prop 52 Analysis

Largest Donors to Prop 52: Medi-Cal Hospital Fee Program

```python
import pandas as pd
```

Reading in the list of propositions and committees from CSV

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head(1)
```

```python
props.info()
```

Read in the list of contributions to ballot measures

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contribs.info()
```

```python
props.prop_name.value_counts().reset_index()
```

```python
props.prop_name.value_counts()
```

```python
prop = props[props.prop_name == "PROPOSITION 052 - STATE FEES ON HOSPITALS. FEDERAL MEDI-CAL MATCHING FUNDS. INITIATIVE STATUTORY AND CONSTITUTIONAL AMENDMENT."]
```

```python
prop.head()
```

```python
prop.info()
```

```python
contribs.info()
```

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum()/merged.amount.sum()
```

```python
merged.sort_values("amount", ascending=False)
```

```python
oppose.sort_values("amount", ascending=False).head(5)
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby("contributor_lastname").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
